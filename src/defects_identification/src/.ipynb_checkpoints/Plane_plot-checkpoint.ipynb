```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm



fig = plt.figure(num=1, clear=True, figsize=[14,14])
ax = fig.add_subplot(1, 1, 1, projection='3d')

(x, y) = np.meshgrid(np.arange(-2, 2.1, 1), np.arange(-1, 1.1, .25))
z1 = (0.145286*x -0.701596*y -0.0719932)/0.697607
z2 = (-0.26805*x +0.364759*y +0.141021)/0.891684


ax.plot_surface(x, y, z1, cmap=cm.copper)
ax.plot_surface(x, y, z2, cmap=cm.copper)
ax.set(xlabel='x', ylabel='y', zlabel='z', title='z')

fig.tight_layout()
```
