## Data visualization

### position, speed

```python
fig = go.Figure()

# Plot points, using distance FROM plane as color scale
fig.add_trace(go.Scatter3d(x = data['X'],
                           y = data['Y'],
                           z = data['Z'],
                           mode='markers',
                           marker=dict(size=5,
                                       color=data['Speed'] [data['Speed'] > 0.0001], #data_meltpool['minor_axis']/5 or data_velocity['speed']
                                       colorscale='oranges',# other colorscale options: agsunset
                                       cmin = 0,  # minimum color value, --- data_velocity['speed'].min()   ### need some adjustment here
                                       cmax = 29,  # maximum color value --0.006 //data['Speed'].max()
                                       showscale=True)
                        )
            )

fig.update_layout(
#     xaxis_title="X Axis",
#     yaxis_title="Y Axis",
#     zaxis_title="Z Axis",
    title={
        'text': "Multidimensional Data visualisation - Colour (Velocity)",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}
    
#     font=dict(
#         family="Courier New, monospace",
#         size=11,
#         color="RebeccaPurple"
#     )
)

fig.update_layout(scene = dict(zaxis = dict(nticks=3,range=[0,10]),
#                                yaxis = dict(nticks=6, range=[data_position_2['y'].min(), data_position_2['y'].max()]),
#                                zaxis = dict(nticks=3, range=[data_position_2['z'].min(), data_position_2['z'].max()]),
#                                aspectmode = 'data'  # preserve the proportion of actual axes data
                               )
                  )


# Use the offline mode of plotly:
pio.write_html(fig, file="data_visualization_velocity2.html", auto_open=True)
```

## Data Synchornization (Current Solution)
- The sampling frequency of speed/position is 250 Hz
- The sampling frequency of melt pool size is 10 Hz
- Repeat each element of the melt pool size 25 times

```python
mps = np.repeat(data['pixels'].to_list(),25)
mps = mps[~np.isnan(mps)]
mps
```

```python
mps_df = pd.DataFrame(mps, columns=['mps'])
```
