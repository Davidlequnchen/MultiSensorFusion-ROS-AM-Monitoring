```python
## short-time fouier transform
X = librosa.stft(x_normal)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr_normal, x_axis='time', y_axis='log')
plt.title("Spectrogram - printing sound- log scale")
plt.colorbar()
```
